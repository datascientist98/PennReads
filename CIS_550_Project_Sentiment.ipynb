```python
df_rating = pd.read_csv('Books_rating.csv')
df_data = pd.read_csv('books_data.csv')
```

```python
df_rating.head()
```

```python
df_data.head()
```

```python
v_sia = SentimentIntensityAnalyzer()

def assign_score(row):
    if v_sia.polarity_scores(row)['compound']<0:
        label = 0 #negative
    else:
        label = 1 #positive 
    return label
```

```python
df_merge = df_data.merge(df_rating, how="inner", on="Title")
```

```python
df_merge = df_merge.sort_values(by='review/helpfulness', ascending=False)
```

```python
df_merge_subset = df_merge.copy()[:1000]
```

```python
df_merge_subset['sentiment'] = df_merge_subset['review/summary'].apply(assign_score)
df_merge_subset
```

```python
df_merge_subset.to_csv('sentiment.csv', encoding='utf-8', index=False, header=True)
```
